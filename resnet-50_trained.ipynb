```python
import keras 
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
from tensorflow.keras.models import load_model
model=load_model('/Users/yeshnavya/Documents/archive_resnet/resnet_30_shallow.h5')
model.summary()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

test_data_dir = '/Users/yeshnavya/Documents/archive_resnet/train'  

#This is without rescaling the data
test_datagen= ImageDataGenerator(rescale= 255/255)

test_generator = test_datagen.flow_from_directory(
    test_data_dir,
    target_size=(64, 64),  
    batch_size=10, 
    class_mode='categorical', 
    shuffle=False
)
```

```python
results = model.evaluate(test_generator)
print(f"Test Loss: {results[0]}, Test Accuracy: {results[1]}")
```
